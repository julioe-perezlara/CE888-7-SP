```python
import pandas as pd # Allow us to work with CSV files
```

```python
# Load Twitter Dataset

df_sent_train = pd.read_csv("https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/sentiment/train_text.txt", \
               sep = '\n', header=None, names=['tweets'])
df_sent_test = pd.read_csv("https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/sentiment/test_text.txt", \
               sep = '\n', header=None, names=['tweets']) 
df_sent_val = pd.read_csv("https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/sentiment/val_text.txt", \
               sep = '\n', header=None, names=['tweets']) 

# Read a CSV file from internet (Github) that does not have a header
```

```python
pd.reset_option('^display.', silent=True) # Reset the default Pandas Display (Truncated)
display(df_sent_train)
display(df_sent_test)
display(df_sent_val)
```

```python
# Comparing Original text vs Clean text (text after preprocessing)
pd.set_option("display.max_rows", None, "display.max_columns", None, 'display.max_colwidth', None)
# This option is used to print the entire Pandas dataframe (all rows, all columns & all content)
df_sent_train[0:10] # Output: Document located at index 10. 
```
